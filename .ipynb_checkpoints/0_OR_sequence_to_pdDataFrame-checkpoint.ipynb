### Goal of this notebook is to read in all the OR sequences and split into pandas data table in order for downstream onehotencoding or splitting 

```python
OR_files = os.listdir('./OR_sequence/')
OR_files
```

```python
from Bio import SeqIO
```

```python
mouseOR_60p = SeqIO.parse(open('./OR_sequence/mouseOR_60p_aaIdentity.fasta'), 'fasta')
OR_sequence = {}
for OR in mouseOR_60p:
    OR_sequence[OR.id] = str(OR.seq)

# print(OR_sequence['Olfr102'])
```

```python
# Creates a OR_table that holds the sequence length of amino acid in individual columns
OR_table = pd.DataFrame()
for aa in range(len(list(OR_sequence.values())[0])):
    OR_table['AA_'+str(aa+1)] =  ""

# Adds OR names as index
OR_table['OR'] = list(OR_sequence.keys())
OR_table = OR_table.set_index('OR')
# Adds in the individual sequence into the OR_table with the correct position 
for OR in OR_table.index:
    OR_table.loc[OR] = list(OR_sequence[OR])
```
